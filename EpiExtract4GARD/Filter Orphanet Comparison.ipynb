# Goals:
Filter the Orphanet complete file to only include:
 - Rows with exactlly 1 stat AND 1 or less GARD IDs

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Orphanet-Comparison-FINAL.csv')
df
```

```python
#converts string sets or lists into a python list
def str2list(string,lessthan):
    if string =='[]':
        #removes all stats with empty lists (list of len 0)
        return np.NaN
    elif string =='set()':
        return []
    else:
        string = str(string).replace("['","")
        string = string.replace("']","")
        string = str(string).replace("{'","")
        string = string.replace("'}","")
        string = string.strip()
        str_list = string.split("', '")
        if len(str_list)>1:
            return np.NaN
        else: 
            return str_list
```

```python
for i in range(len(df['GARD Disease ID'])):
    df['GARD Disease ID'][i] = str2list(df['GARD Disease ID'][i],True)
    df['Epi Statistics'][i] = str2list(df['Epi Statistics'][i],False)
```

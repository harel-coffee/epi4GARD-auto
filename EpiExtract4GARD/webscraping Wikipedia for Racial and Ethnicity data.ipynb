## Webscraping Wikipedia articles for racial and ethnic terms

```python
#import sys
#!{sys.executable} -m pip install bs4
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
import re
```

```python
ethnicities = set() #all names of ethnicities in string set
```

## List of contemporary ethnic groups

```python
url = 'https://en.wikipedia.org/wiki/List_of_contemporary_ethnic_groups'
```

```python
html = urlopen(url) 
soup = BeautifulSoup(html, 'html.parser')
tables = soup.find_all('table')
```

```python
for table in tables:
    rows = table.find_all('tr')
    
    for row in rows:
        cells = row.find_all('td')
        
        
        if len(cells) > 1:
            ethnicity = cells[0]
            ethnicity = ethnicity.text.split('\n')
            if len(ethnicity)<=2:
                print(ethnicity[0])
                ethnicities.add(ethnicity[0].strip())
```

```python
len(ethnicities)
```

## South Asian ethnic groups

```python
url = 'https://en.wikipedia.org/wiki/South_Asian_ethnic_groups'
```

```python
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
ulist = soup.find_all('ul')
```

```python
for i in range(4,53):
    list_entries = ulist[i].find_all('li')
    for entry in list_entries:
        ethnicity = re.sub('\[.\]', '', entry.text)
        ethnicity = ethnicity.strip().split()
        if 'people' in ethnicity or 'in' in ethnicity:
            print(ethnicity[0])
            ethnicities.add(ethnicity[0])
        elif len(ethnicity)<3:
            print(' '.join(ethnicity))
            ethnicities.add(' '.join(ethnicity))
```

## List of ethnic groups in Myanmar

```python
url = 'https://en.wikipedia.org/wiki/List_of_ethnic_groups_in_Myanmar'
```

```python
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
ulist = soup.find_all('ul')
```

## Process the names

```python
import nltk
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize('Italians'))

from nltk.stem import PorterStemmer
porter = PorterStemmer()
print(porter.stem('Italians'))
print(porter.stem('Malayali'))
print(porter.stem('Chinese'))
print(porter.stem('Indian'))
print(porter.stem('South'))
print(porter.stem('Marie'))
print(porter.stem('SARA'))
print(porter.stem('LOM'))
#Going to only use the stemmer

from nltk.corpus import stopwords
#nltk.download('stopwords')
STOPWORDS = set(stopwords.words('english'))
import string
PUNCTUATION = set(char for char in string.punctuation)
#The second list was generated from looking at the modify labelsV4  ipynb
STOPWORDS = STOPWORDS.union(PUNCTUATION,{'South','south','Marie','marie','mari','even','raw','Iron','iron','pale','tongue','Tongue','far','mixed','SARA','sara','ironically','AMI','ami','anus','fang','LOM','Lom','lom','Pars',porter.stem('Tongue')})
STOPWORDS = STOPWORDS.union({porter.stem(word) for word in STOPWORDS})
```

```python
print(len(ethnicities))
ethnicities = {ethnicity for ethnicity in ethnicities if len(ethnicity)>2 and ethnicity not in STOPWORDS}
ethnicities.discard('Botswana')
ethnicities.add('Batswana')
ethnicities.add('Motswana')
print(len(ethnicities))
```

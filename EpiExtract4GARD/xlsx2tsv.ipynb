```python
import openpyxl
import csv
```

```python
wb = openpyxl.load_workbook('Epi4GARD_test_set[CORRECTED].xlsx')
sheet = wb['Test Set']
```

```python
for row in sheet.iter_rows(min_row=2,max_col=2,max_row=14444,values_only=True):
    print(row[0],row[1])
```

```python
tokens,labels = [],[]
sentence_tokens, sentences_labels=[],[]
i=2
for row in sheet.iter_rows(min_row=2,max_col=2,max_row=14445,values_only=True):
    if isinstance(row[0],str) ==False or isinstance(row[1],str)==False:
        print('TYPE ERROR \nROW: ',i)
        print('ROW VALUE: ',row,'\nCOL1: ',row[0],'\nCOL2: ',row[1])
    if len(row[0])==0 and len(row[1])==0:
        tokens.append(sentence_tokens.copy())
        labels.append(sentences_labels.copy())
        sentence_tokens.clear()
        sentences_labels.clear()
    elif len(row[0])==0 or len(row[1])==0:
        print('ERROR')
        print(i,row,row[0],row[1])
    else:
        sentence_tokens.append(row[0])
        sentences_labels.append(row[1])
    i+=1
```

```python
#Save as .tsv file
with open('epi_test_setV2-corrected.tsv', "w") as f:
    for i in range(len(tokens)): #for sentence in sentences
        for j in range(len(tokens[i])): #for token in sentence
            output = str(tokens[i][j]) +'\t' +str(labels[i][j])+'\n'
            f.write(output)
        f.write('\n')
        if i%25==0:
            print('abstract num',i,'done')
f.close()
```
